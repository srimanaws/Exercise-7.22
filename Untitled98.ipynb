```python
#Exercise 7.22 by sriman
import pandas as pd
import numpy as np

# a) Let's kick things off by creating a Series from the list [7, 11, 13, 17].
series_a = pd.Series([7, 11, 13, 17])
print("Series A:\n", series_a)

# b) Now, let’s create another Series with five elements, and they’ll all be 100.0.
series_b = pd.Series([100.0] * 5)
print("\nSeries B:\n", series_b)

# c) For this next one, we need a Series with 20 random numbers between 0 and 100.
series_c = pd.Series(np.random.randint(0, 101, size=20))
print("\nSeries C:\n", series_c)

# Now, let's get some basic stats on these random numbers.
print("\nDescriptive Statistics for Series C:\n", series_c.describe())

# d) Next, we're going to create a Series called temperatures. We’ll use some floating-point values and give them custom names.
temperatures = pd.Series([98.6, 98.9, 100.2, 97.9], index=['Julie', 'Charlie', 'Sam', 'Andrea'])
print("\nTemperatures Series:\n", temperatures)

# e) Lastly, let's whip up a dictionary from the names and values we just made. Then, we'll use it to start another Series.
temp_dict = {'Julie': 98.6, 'Charlie': 98.9, 'Sam': 100.2, 'Andrea': 97.9}
series_e = pd.Series(temp_dict)
print("\nSeries E from Dictionary:\n", series_e)
```
